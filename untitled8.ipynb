```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc
from pyspark.sql.types import (StructType, StructField, StringType,
                               DoubleType, IntegerType, LongType)

spark = SparkSession.builder.appName("NCR Ride Analysis").getOrCreate()


schema = StructType([
    StructField("Date", StringType(), True),
    StructField("Time", StringType(), True),
    StructField("Booking ID", StringType(), True),
    StructField("Booking Status", StringType(), True),
    StructField("Customer ID", StringType(), True),
    StructField("Vehicle Type", StringType(), True),
    StructField("Pickup Location", StringType(), True),
    StructField("Drop Location", StringType(), True),
    StructField("Avg VTAT", DoubleType(), True),
    StructField("Avg CTAT", DoubleType(), True),
    StructField("Cancelled Rides by Customer", IntegerType(), True),
    StructField("Reason for cancelling by Customer", StringType(), True),
    StructField("Cancelled Rides by Driver", IntegerType(), True),
    StructField("Driver Cancellation Reason", StringType(), True),
    StructField("Incomplete Rides", IntegerType(), True),
    StructField("Incomplete Rides Reason", StringType(), True),
    StructField("Booking Value", DoubleType(), True),
    StructField("Ride Distance", DoubleType(), True),
    StructField("Driver Ratings", DoubleType(), True),
    StructField("Customer Rating", DoubleType(), True),
    StructField("Payment Method", StringType(), True),
])

csv_path = "ncr_ride_bookings.csv"


df = spark.read.csv(csv_path, header=True, schema=schema)


top_pickup_locations = df.groupBy("Pickup Location") \
                         .count() \
                         .orderBy(desc("count")) \
                         .limit(10)

top_drop_locations = df.groupBy("Drop Location") \
                       .count() \
                       .orderBy(desc("count")) \
                       .limit(10)

print("--- Top 10 Busiest Pickup Locations ---")
top_pickup_locations.show(truncate=False)

print("\n--- Top 10 Busiest Drop Locations ---")
top_drop_locations.show(truncate=False)


spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, sum, avg, desc
from pyspark.sql.types import (StructType, StructField, StringType,
                               DoubleType, IntegerType, LongType)


spark = SparkSession.builder.appName("NCR Ride Analysis").getOrCreate()


schema = StructType([
    StructField("Date", StringType(), True),
    StructField("Time", StringType(), True),
    StructField("Booking ID", StringType(), True),
    StructField("Booking Status", StringType(), True),
    StructField("Customer ID", StringType(), True),
    StructField("Vehicle Type", StringType(), True),
    StructField("Pickup Location", StringType(), True),
    StructField("Drop Location", StringType(), True),
    StructField("Avg VTAT", DoubleType(), True),
    StructField("Avg CTAT", DoubleType(), True),
    StructField("Cancelled Rides by Customer", IntegerType(), True),
    StructField("Reason for cancelling by Customer", StringType(), True),
    StructField("Cancelled Rides by Driver", IntegerType(), True),
    StructField("Driver Cancellation Reason", StringType(), True),
    StructField("Incomplete Rides", IntegerType(), True),
    StructField("Incomplete Rides Reason", StringType(), True),
    StructField("Booking Value", DoubleType(), True),
    StructField("Ride Distance", DoubleType(), True),
    StructField("Driver Ratings", DoubleType(), True),
    StructField("Customer Rating", DoubleType(), True),
    StructField("Payment Method", StringType(), True),
])


csv_path = "ncr_ride_bookings.csv"


df = spark.read.csv(csv_path, header=True, schema=schema)




df_filtered = df.filter(col("Pickup Location").isNotNull() & col("Booking Value").isNotNull())


city_analysis = df_filtered.groupBy("Pickup Location") \
    .agg(
        count("*").alias("Total Rides"),
        sum("Booking Value").alias("Total Revenue"),
        avg("Booking Value").alias("Average Fare Per Trip")
    ) \
    .orderBy(desc("Total Rides"))



print("--- City-Wise Ride Analysis ---")
city_analysis.show(20, truncate=False)


spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, hour, when, desc, to_timestamp
from pyspark.sql.types import (StructType, StructField, StringType,
                               DoubleType, IntegerType, LongType)


spark = SparkSession.builder.appName("NCR Ride Analysis").getOrCreate()


schema = StructType([
    StructField("Date", StringType(), True),
    StructField("Time", StringType(), True),
    StructField("Booking ID", StringType(), True),
    StructField("Booking Status", StringType(), True),
    StructField("Customer ID", StringType(), True),
    StructField("Vehicle Type", StringType(), True),
    StructField("Pickup Location", StringType(), True),
    StructField("Drop Location", StringType(), True),
    StructField("Avg VTAT", DoubleType(), True),
    StructField("Avg CTAT", DoubleType(), True),
    StructField("Cancelled Rides by Customer", IntegerType(), True),
    StructField("Reason for cancelling by Customer", StringType(), True),
    StructField("Cancelled Rides by Driver", IntegerType(), True),
    StructField("Driver Cancellation Reason", StringType(), True),
    StructField("Incomplete Rides", IntegerType(), True),
    StructField("Incomplete Rides Reason", StringType(), True),
    StructField("Booking Value", DoubleType(), True),
    StructField("Ride Distance", DoubleType(), True),
    StructField("Driver Ratings", DoubleType(), True),
    StructField("Customer Rating", DoubleType(), True),
    StructField("Payment Method", StringType(), True),
])


csv_path = "ncr_ride_bookings.csv"

df = spark.read.csv(csv_path, header=True, schema=schema)


df_with_hour = df.withColumn("hour", hour(to_timestamp(col("Time"), "HH:mm:ss")))


df_with_time_slots = df_with_hour.withColumn(
    "Time Slot",
    when((col("hour") >= 5) & (col("hour") < 8), "Early Morning (5-8 AM)")
    .when((col("hour") >= 8) & (col("hour") < 12), "Morning (8-12 PM)")
    .when((col("hour") >= 12) & (col("hour") < 16), "Afternoon (12-4 PM)")
    .when((col("hour") >= 16) & (col("hour") < 19), "Evening (4-7 PM)")
    .when((col("hour") >= 19) & (col("hour") < 22), "Night (7-10 PM)")
    .otherwise("Late Night")
)


peak_hours_analysis = df_with_time_slots.groupBy("Time Slot") \
    .count() \
    .orderBy(desc("count"))



print("--- Peak Ride Hours Analysis ---")
peak_hours_analysis.show(truncate=False)


spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when
import matplotlib.pyplot as plt
import seaborn as sns


spark = SparkSession.builder.appName("NCR Ride Analysis").getOrCreate()


df = spark.read.csv("ncr_ride_bookings.csv", header=True, inferSchema=True)


df_payment_category = df.withColumn(
    "Payment Category",
    when(col("Payment Method") == "Cash", "Cash")
    .when((col("Payment Method") == "UPI") |
          (col("Payment Method") == "Debit Card") |
          (col("Payment Method") == "Credit Card"), "Online")
    .otherwise("Other")
)

payment_share = df_payment_category \
    .filter(col("Payment Category").isin("Cash", "Online")) \
    .groupBy("Payment Category") \
    .count()


payment_data = payment_share.collect()
labels = [row["Payment Category"] for row in payment_data]
sizes = [row["count"] for row in payment_data]


plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140,
        colors=sns.color_palette("pastel"))
plt.title("Share of Cash vs. Online Payments")
plt.ylabel("")


plt.savefig("payment_share_pie_chart.png")



print("--- Payment Method Share ---")
payment_share.show()


spark.stop()
```
